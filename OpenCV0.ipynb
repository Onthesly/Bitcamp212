```python
# 비디오 출력
cap = cv2.VideoCapture(0)

while(True):
    ret, frame = cap.read()
    src = cv2.cvtColor(frame, cv2.IMREAD_COLOR)
    dst = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    cv2.imshow('SRC', src)
    cv2.imshow('DST', dst)
    if cv2.waitKey(1) & 0xff == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
